```python
import os
import fnmatch

class GLOB:
    def __init__(self, path):
        self.path = path

    def glob(self, pattern):
        return self._find_matches(self.path, pattern)

    def iglob(self, pattern):
        return self._iter_matches(self.path, pattern)

    def _find_matches(self, path, pattern):
        matches = []
        for root, dirs, files in os.walk(path):
            for file in files:
                if self._match_pattern(file, pattern):
                    matches.append(os.path.join(root, file))
        return matches

    def _iter_matches(self, path, pattern):
        for root, dirs, files in os.walk(path):
            for file in files:
                if self._match_pattern(file, pattern):
                    yield os.path.join(root, file)

    @staticmethod
    def _match_pattern(file, pattern):
        import fnmatch
        return fnmatch.fnmatch(file, pattern)

# Usage example:
glob_obj = GLOB(r"C:\Users\HITESH\Desktop\New folder")

# Find all .txt files
txt_files = glob_obj.glob('*.txt')
print(txt_files)

# Find all .py files iteratively
for py_file in glob_obj.iglob('*.py'):
    print(py_file)
# find all ipynb files
for ipynb_file in glob_obj.iglob('*.ipynb'):
    print(ipynb_file)
# find all php files
for php_file in glob_obj.iglob('*.php'):
    print(php_file)
# find all js files
for js_file in glob_obj.iglob('*.js'):
    print(js_file)
# find all html files
for html_file in glob_obj.iglob('*.html'):
    print(html_file)
# find all mp4 files
for mp4_file in glob_obj.iglob('*.mp4'):
    print(mp4_file)
# find all mp3 files
for mp3_file in glob_obj.iglob('*.mp3'):
    print(mp3_file)
# find all jpg files
for jpg_file in glob_obj.iglob('*.jpg'):
    print(jpg_file)
# find all jpeg files
for jpeg_file in glob_obj.iglob('*.jpeg'):
    print(jpeg_file)
# find all jpeg files
for png_file in glob_obj.iglob('*.png'):
    print(png_file)
# find all pdf files
for pdf_file in glob_obj.iglob('*.pdf'):
    print(pdf_file)
# find all docx files
for docx_file in glob_obj.iglob('*.docx'):
    print(docx_file)
```
